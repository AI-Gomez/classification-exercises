```python
def new_iris_data():
    sql_query = """
                SELECT species_id,
                species_name,
                sepal_length,
                sepal_width,
                petal_length,
                petal_width
                FROM measurements
                JOIN species
                USING(species_id)
                """
    df = pd.read_sql(sql_query, get_connection('iris_db'))
    df.to_csv('iris_df.csv')
    return df

def get_iris_data(cached=False):
    if cached or os.path.isfile('iris_df.csv') == False:
        df = new_iris_data()
    else:
        df = pd.read_csv('iris_df.csv', index_col=0)
    return df
```

```python
#function to do it all
def titanic_split(df):
    train_validate, test = train_test_split(df, test_size=.2, 
                                        random_state=123, 
                                        stratify=df.survived)
    train, validate = train_test_split(train_validate, test_size=.3, 
                                   random_state=123, 
                                   stratify=train_validate.survived)
    return train, validate, test
```

```python
# contain it all in a function

def impute_mean_age(train, validate, test):
    imputer = SimpleImputer(strategy = 'mean')
    train['age'] = imputer.fit_transform(train[['age']])    
    validate['age'] = imputer.transform(validate[['age']])   
    test['age'] = imputer.transform(test[['age']])
    return train, validate, test
```

```python
#Create a function named prep_titanic that accepts the untransformed 
#titanic data, and returns the data with the transformations above applied.
def prep_titanic(cached=True):
    df = get_titanic_data(cached)
    df = df[~df.embarked.isnull()]
    titanic_dummies = pd.get_dummies(df.embarked, drop_first=True)
    df = pd.concat([df, titanic_dummies], axis=1)
    df = df.drop(columns='deck')
    train, validate, test = titanic_split(df)
    train, validate, test = impute_mean_age(train, validate, test)   
    return train, validate, test

train, validate, test = prep_titanic()
```

```python
#Create 4 subplots (2 rows x 2 columns) of scatterplots.
#sepal_length x sepal_width
#petal_length x petal_width
#sepal_area x petal_area
#sepal_length x petal_length

train['sepal_area'] = train.sepal_length * train.sepal_width
train['petal_area'] = train.petal_length * train.petal_width
```

```python
def scatter_by_species(x, y):
    for species, subset in train.groupby('species'):
        plt.scatter(subset[x], subset[y], label=species)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend()

scatter_by_species('sepal_length', 'sepal_width')
```

```python
#Is the sepal length significantly different in virginica 
#compared to versicolor?
#Make sure to include a null hypothesis, alternative hypothesis,
#results, and summary.

#alpha = .05

# Ho: sepal length is the same in virginica and versicolor
# Ha: sepal length significantly different in virginica and versicolor

virginica = train[train.species == 'virginica']
versicolor = train[train.species == 'versicolor']
```

```python
train.sepal_length.describe()
```

```python
virginica.sepal_length.var(), versicolor.sepal_length.var()
```

```python
import scipy.stats as stats


t, p = stats.ttest_ind(
    virginica.sepal_length,
    versicolor.sepal_length,
)
print('t=', t, 'p=', p)
```
